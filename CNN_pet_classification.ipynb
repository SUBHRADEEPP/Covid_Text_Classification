```python
#importing necessary libraries

from glob2 import glob #storage

from tensorflow.keras import Sequential
from tensorflow.keras.layers import Conv2D,Dense,Dropout,Flatten,MaxPooling2D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import categorical_crossentropy
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
#defining the path for train and test dataset 

train_path = '/content/data/train'
test_path = '/content/data/test'
```

```python
images_train = glob(train_path+"/*/*.jpg")
images_test = glob(test_path+"/*/*.jpg")
```

```python
generator = ImageDataGenerator()
```

```python
train_gen = generator.flow_from_directory(train_path)
```

```python
test_gen = generator.flow_from_directory(test_path)
```

```python
#Defining CNN model

model=Sequential()
model.add(Conv2D(32,(5,5),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2),strides=2))
model.add(Conv2D(64,kernel_size=5,activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2),strides=2))
model.add(Flatten())
model.add(Dense(32,activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(2, activation='softmax'))
```

```python
#Compiling CNN model

model.compile(optimizer='Adam',loss ='categorical_crossentropy',metrics=['accuracy'])
```

```python
#Fitting the data in the model 
#epochs =100

model.fit(train_gen,epochs=100, validation_data=test_gen)
```

```python
#epochs=200

model.fit(train_gen,epochs=200, validation_data=test_gen)
```

```python
#epochs 300

model.fit(train_gen,epochs=300, validation_data=test_gen)
```
